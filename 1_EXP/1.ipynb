```python
"""**Data Formatting and Data Normalization: Summarize the types of variables by checking the
data types (i.e., character, numeric, integer, factor, and logical) of the variables in the data set.
If variables are not in the correct data type, apply proper type conversions.**
"""

# Checking data types
df.dtypes

```

```python
# Convert Survived column into category
df['Survived'] = df['Survived'].astype('category')

```

```python
# Convert Sex column into category
df['Sex'] = df['Sex'].astype('category')

```

```python
# Checking updated data types
df.dtypes

```

```python
# Normalizing Fare column
df['Fare_Normalized'] = (
    (df['Fare'] - df['Fare'].min()) /
    (df['Fare'].max() - df['Fare'].min()))
df[['Fare', 'Fare_Normalized']].head()

"""**Turn categorical variables into quantitative variables in Python.**"""

```

```python
# Convert Sex column into numeric values
df['Sex'] = df['Sex'].map({
    'male': 0,
    'female': 1
})
df[['Sex']].head()

```

```python
# One-hot encoding
df = pd.get_dummies(df, columns=['Embarked'])
df.head()

df.info()
```
