# Descriptive Statistics - Measures of Central Tendency and variability
Perform the following operations on any open source dataset (e.g., data.csv) 1. Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative variables. 2. Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Irisversicolor’ and ‘Iris- versicolor’ of iris.csv dataset.

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/iris.csv")

# IMPORTANT FIX: Your code expects columns like 'Species' and 'SepalLengthCm', 
# but the uploaded iris.csv uses 'class' and 'sepal length (cm)'. 
# We rename them here so the rest of your original code works perfectly!
df.rename(columns={
    'sepal length (cm)': 'SepalLengthCm',
    'sepal width (cm)': 'SepalWidthCm',
    'petal length (cm)': 'PetalLengthCm',
    'petal width (cm)': 'PetalWidthCm',
    'class': 'Species'
}, inplace=True)

if 'target' in df.columns:
    df.drop('target', axis=1, inplace=True)
if 'Id' not in df.columns:
    df.insert(0, 'Id', range(1, 1 + len(df)))

df['Species'] = df['Species'].replace({
    'setosa': 'Iris-setosa', 
    'versicolor': 'Iris-versicolor', 
    'virginica': 'Iris-virginica'
})

df.head()
```

**Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset with numeric variables grouped by one of the qualitative (categorical) variable.**

```python
# Mean Grouped by Species
df.groupby('Species').mean()
```

```python
# Median Grouped by Species
df.groupby('Species').median()
```

```python
# Minimum Values Grouped by Species
df.groupby('Species').min()
```

```python
# Maximum Values Grouped by Species
df.groupby('Species').max()
```

```python
# Standard Deviation Grouped by Species
df.groupby('Species').std()
```

**Create a list that contains a numeric value for each response to the categorical variable.**

```python
# Sepal Length values for each species
setosa = df[df['Species'] == 'Iris-setosa']['SepalLengthCm'].tolist()
versicolor = df[df['Species'] == 'Iris-versicolor']['SepalLengthCm'].tolist()
virginica = df[df['Species'] == 'Iris-virginica']['SepalLengthCm'].tolist()

print("Setosa:", setosa)
print("Versicolor:", versicolor)
print("Virginica:", virginica)
```

**Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-virginica’ of iris.csv dataset.**

```python
# Filter Species Data
setosa = df[df['Species'] == 'Iris-setosa']
versicolor = df[df['Species'] == 'Iris-versicolor']
virginica = df[df['Species'] == 'Iris-virginica']
```

```python
# Statistical Details of Iris-setosa
print("Iris-setosa Statistics")
display(setosa.describe())
```

```python
# Statistical Details of Iris-versicolor
print("Iris-versicolor Statistics")
display(versicolor.describe())
```

```python
# Statistical Details of Iris-virginica
print("Iris-virginica Statistics")
display(virginica.describe())
```

**Percentile Calculation**

```python
print("Setosa Percentiles")
print(np.percentile(setosa['SepalLengthCm'], [25, 50, 75]))

print("Versicolor Percentiles")
print(np.percentile(versicolor['SepalLengthCm'], [25, 50, 75]))

print("Virginica Percentiles")
print(np.percentile(virginica['SepalLengthCm'], [25, 50, 75]))
```
