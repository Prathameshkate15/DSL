```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Load your uploaded iris.csv file
df = pd.read_csv('/content/iris.csv')

# IMPORTANT FIX: Your CSV has different column names than the PDF expects.
# We rename them here so the rest of the code works perfectly!
df.rename(columns={
    'sepal length (cm)': 'SepalLengthCm',
    'sepal width (cm)': 'SepalWidthCm',
    'petal length (cm)': 'PetalLengthCm',
    'petal width (cm)': 'PetalWidthCm',
    'class': 'Species'
}, inplace=True)

if 'Id' not in df.columns:
    df.insert(0, 'Id', range(1, 1 + len(df)))
df['Species'] = df['Species'].replace({'setosa': 'Iris-setosa', 'versicolor': 'Iris-versicolor', 'virginica': 'Iris-virginica'})

df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))

axes[0, 0].set_title("Distribution of first column")
axes[0, 0].hist(df["SepalLengthCm"])

axes[0, 1].set_title("Distribution of second column")
axes[0, 1].hist(df["SepalWidthCm"])

axes[1, 0].set_title("Distribution of Third Column")
axes[1, 0].hist(df["PetalLengthCm"])

axes[1, 1].set_title("Distribution of Fourth Column")
axes[1, 1].hist(df["PetalWidthCm"])

plt.show()
```

```python
sns.set_style("whitegrid")
data_to_plot = [df["SepalLengthCm"], df["SepalWidthCm"], df["PetalLengthCm"], df["PetalWidthCm"]]

# Creating a figure instance
fig = plt.figure(1, figsize=(12, 8))

# Creating an axes instance
ax = fig.add_subplot(111)

# Creating the boxplot
bp = ax.boxplot(data_to_plot)
plt.show()
```
