# Text Analytics and NLP Basics
Tokenization, POS Tagging, Stopword Removal, Stemming, Lemmatization, and TF-IDF.

```python
import nltk
import string
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# Download required NLTK datasets
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
document = """Text analytics is the process of analyzing textual data to extract meaningful insights.
It includes preprocessing steps like tokenization, stop word removal, stemming, and lemmatization."""

print("Original Document:\n", document)
```

```python

nltk.download('punkt_tab', quiet=True)
tokens = word_tokenize(document)
print("\nTokens:\n", tokens)
```

```python
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
pos_tags = pos_tag(tokens)
print("\nPOS Tags:\n", pos_tags)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word not in string.punctuation]

print("\nAfter Stopword Removal:\n", filtered_tokens)
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nStemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Words:\n", lemmatized_words)
```

```python
vectorizer = CountVectorizer()
tf_matrix = vectorizer.fit_transform([document])
tf_df = pd.DataFrame(tf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print("\nTerm Frequency (TF):")
display(tf_df)
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([document])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

print("\nTF-IDF Representation:")
display(tfidf_df)
```
